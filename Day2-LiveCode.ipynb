```python
import requests
import time
import pathlib
```

```python
target = pathlib.Path('livepictures')
target.mkdir(exist_ok = True)
```

```python
!pwd
```

```python
for u in urls:
    print(u)
    f = pathlib.Path("_".join(u.split("/")[-2:]))
    full_p = target / f
    
    if full_p.exists():
        print("already got it!")
    else:
        print(full_p)
        r = requests.get(u)
        image = r.content
        full_p.write_bytes(image)
        r.close()
        time.sleep(5)
```
